```python
#pandas titanic dataset
import pandas as pd
titanic=pd.read_csv("titanic_lyst7308.csv")
#printing first 10 rows
titanic.head(10)
```

```python
#printing last 10 rows
titanic.tail(10)
```

```python
#filtering passengers older than 25
above_25=titanic[titanic["age"]>25]
above_25.head(10)
```

```python
#selecting the name of passengers older the 35 
passengers_name=titanic.loc[titanic["age"]>35,"name"]
passengers_name.head(8)
```

```python
#median age and ticket fare of passengers
titanic[["age","fare"]].median()
```

```python
#min,max and mean of age and fare of passengers using.agg()method
titanic.agg({"age":['min','max','mean'],"fare":['min','max','mean']})
```

```python
#mean ticket fare price foreach sex and cabin class combination
titanic.groupby(["sex","pclass"])["fare"].mean()
```

```python
#sorting titanic data according to age of passengers
titanic.sort_values(by="age").head(10)
```
